```python
df1 = pd.read_csv('country_wise_latest.csv')
df2 = pd.read_csv('worldometer_data.csv')

rename_dict = {
    'USA': 'US',
    'UK': 'United Kingdom',
    'Taiwan': 'Taiwan*',
    'Myanmar': 'Burma',
    'Ivory Coast': "Cote d'Ivoire",
    'Congo': 'Congo (Brazzaville)',
    'DRC': 'Congo (Kinshasa)',
    'Vatican City': 'Holy See',
    'Palestine': 'West Bank and Gaza',
    'S. Korea': 'South Korea',
    'CAR': 'Central African Republic',
    'Brunei ': 'Brunei',
}

df2['Country/Region'] = df2['Country/Region'].replace(rename_dict)

merged = pd.merge(df1, df2, on='Country/Region', how='outer')

merged
```

```python
merged.to_csv("Final_covid-19.csv", index = False)
```
